### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento


```python
#Soma de faturamento total de cada loja
faturamento_lojas = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

print(faturamento_lojas)
```

```python
#Faturamento em gráfico de pizza
faturamento_lojas = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

labels = faturamento_lojas.keys()
valores = faturamento_lojas.values()

plt.figure(figsize=(8, 6))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Faturamento Total por Loja')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
#Agrupando e somando as vendas por categorias para cada loja
categorias_loja1 = loja1['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()

#Exibindo os resultados
print("Loja 1:")
print(categorias_loja1)
print("\nLoja 2:")
print(categorias_loja2)
print("\nLoja 3:")
print(categorias_loja3)
print("\nLoja 4:")
print(categorias_loja4)

```

```python
cat1 = loja1['Categoria do Produto'].value_counts()
cat2 = loja2['Categoria do Produto'].value_counts()
cat3 = loja3['Categoria do Produto'].value_counts()
cat4 = loja4['Categoria do Produto'].value_counts()

df_categorias = pd.DataFrame({
    'Loja 1': cat1,
    'Loja 2': cat2,
    'Loja 3': cat3,
    'Loja 4': cat4
})

df_categorias.plot(kind='bar', figsize=(10, 6))

plt.title('Vendas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend(title='Lojas')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
#Calculando a média das avaliações de cada loja
avaliacoes = {
    'Loja 1': round(loja1['Avaliação da compra'].mean(), 2),
    'Loja 2': round(loja2['Avaliação da compra'].mean(), 2),
    'Loja 3': round(loja3['Avaliação da compra'].mean(), 2),
    'Loja 4': round(loja4['Avaliação da compra'].mean(), 2)
}
#Exibindo as médias
print(avaliacoes)
```

# 4. Produtos Mais e Menos Vendidos

```python
#Função para identificar os produtos mais e menos vendidos
def mais_menos_vendidos(df, nome_loja):
  contagem = df['Produto'].value_counts()
  mais = contagem.idxmax()
  qtd_mais = contagem.max()
  menos = contagem.idxmin()
  qtd_menos = contagem.min()

  return {
      'Loja': nome_loja,
      'Mais vendido': f'{mais} ({qtd_mais})',
      'Menos vendido': f'{menos} ({qtd_menos})'
  }

#aplicando para todas as lojas
resultado = [
  mais_menos_vendidos(loja1, 'Loja 1'),
  mais_menos_vendidos(loja2, 'Loja 2'),
  mais_menos_vendidos(loja3, 'Loja 3'),
  mais_menos_vendidos(loja4, 'Loja 4')
]

#Transformando em DataFrame para visualização
df_resultado = pd.DataFrame(resultado)
print(df_resultado)
```

```python
#gráfico de dispersão
x = range(len(df_resultado))
lojas = df_resultado['Loja']

mais = [int(valor.split('(')[1].replace(')', '')) for valor in df_resultado['Mais vendido']]
nome_mais = [valor.split(' (')[0] for valor in df_resultado['Mais vendido']]

menos = [int(valor.split('(')[1].replace(')', '')) for valor in df_resultado['Menos vendido']]
nome_menos = [valor.split(' (')[0] for valor in df_resultado['Menos vendido']]

plt.figure(figsize=(10,6))
plt.scatter(x, mais, color='green', label='Mais Vendido', s=100)
plt.scatter(x, menos, color='red', label='Menos Vendido', s=100)

for i in range(len(lojas)):
  plt.annotate(nome_mais[i], (x[i], mais[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='green')
  plt.annotate(nome_menos[i], (x[i], menos[i]), textcoords="offset points", xytext=(0,-15), ha='center', fontsize=8, color='red')

plt.xticks(x, lojas)
plt.ylabel('Quantidade de Vendas')
plt.title('Produto Mais e Menos Vendido por Loja')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
#Calculando o frete médio de cada Loja
frete_medio = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

for loja, frete in frete_medio.items():
  print(f'{loja}: R$ {frete:.2f}')
```

Com base na análise de desempenho das quatro lojas, a recomendação é que o Senhor João venda a Loja 4. Apesar de ter o menor frete médio, essa unidade apresenta o menor faturamento, média representatividade nas categorias mais vendidas. Em comparação com as demais, é a loja com menor eficiência geral e, portanto, a melhor candidata para ser vendida.
